## Function

Create a function which will generate a calendar.

The parameter have to be a Year, the out a Dictionnary with the following keys.

The expected result is : {'day': [1, 2...], 'month': ['January', 'January'...'December'], 'year': [2020, 2020..], 'elapsed_time': [0.003, ...100]}

```python
import calendar
import time


def new_calendar(year):
    new_time = time.time()

    dictio = {'day': [], 'month': [], 'year': [], 'elapsed_time': []}

    for month in range(1, 13):
        for day in range(1, calendar.monthrange(year, month)[1] + 1):
            dictio['day'].append(day)
            dictio['month'].append(calendar.month_name[month])
            dictio['year'].append(year)

    elapsed_time = time.time() - new_time
    dictio['elapsed_time'] = [elapsed_time] * len(dictio['day'])

    return dictio

year_chosen = 2024
myCalendar = new_calendar(year_chosen)
print(myCalendar)
```
