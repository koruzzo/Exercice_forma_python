# Loop

By using a while loop, create a list of numbers from 1 to 31 included.

```python
list = []
i = 1
while i < 32:
    list.append(i)
    i+=1
print(list)
```

By using a for loop, create a list of tuple, containing a number and a month.
All months can have 31 days.

Expected result : [(1, 'January'), (2, 'January'), (31, 'DÃ©cembre')]

```python
name_month = ["January", "February", "March", "April", "May", "June","July", "August", "September", "October", "November", "December"]

list_days_and_months = []

for month in range(1, 13):
    list_days_and_months.extend([(day, name_month[month-1]) for day in range(1, 32)])
for item in list_days_and_months:
    print(item)
```
